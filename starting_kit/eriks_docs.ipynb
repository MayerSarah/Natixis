```python
'''

hexuity.data_expander unpacks the data semi-nicely

parameters:

    index : 'vix' or 'eurusd', default = 'vix'
        gets data for selected index

    task : 'classification' or 'regression', default 'classification
        returns series for selected target variable

    test_size= : [0,1], default = 0
        splits data into train and test data

    price_data : boolean, default = True
        decides whether to include price_data 
    
    speech_data : boolean, default = True
        decides whether to include speech_data

    all_data : boolean, default = False
        returns all target and input data

    validation : boolean, default = False
        returns otherwise hidden validation data

returns:

    X and y

    if test_size > 0:
        X_train, X_test, y_train, y_test

    if all_data:
        X, y_classification, y_regression

    if validation:
        X_train, X_val, y_train, y_val
    
'''

from hexuity import data_expander
```

```python
X, y = data_expander(index='vix')
```

```python
from lightgbm import LGBMClassifier
from sklearn.metrics import accuracy_score

X_train, X_test, y_train, y_test = data_expander(
    index='eurusd', task='classification',
    test_size=0.2, speech_data=False)

classifer = LGBMClassifier()
classifer.fit(X_train, y_train)
preds = classifer.predict(X_test)

accuracy_score(preds, y_test)
```

```python
'''

hexuity.metrics allows for easy testing of models (sklearn API dependant)

parameters:

    vixclassifier : classifier, default = None
        requires sklearn api

    vixregressor : regressor, default = None
        requires sklearn api

    eurclassifier : classifier, default = None
        if none : same as vix

    eurregressor : regressor, default = None
        if none : same as vix

    test_split= : [0,1], default = 0.2
        splits data into train and test data

    prices : boolean, default = True
        decides whether to include price data 
    
    speeches : boolean, default = True
        decides whether to include speech data

    valid : boolean, default = False
        uses validation data

returns:

    accuracy and rmse for both indices 

'''

from hexuity import metrics
```

```python
from lightgbm import LGBMClassifier, LGBMRegressor
from sklearn.linear_model import LinearRegression, LogisticRegression

vix_classifer = LGBMClassifier()
vix_regressor = LGBMRegressor()

eurusd_classifier = LogisticRegression()
eurusd_regressor = LinearRegression()

metrics(
    vixclassifier=vix_classifer, vixregressor=vix_regressor,
    eurclassifier = eurusd_classifier, eurregressor=eurusd_regressor,
    prices=True, speeches=False,
    test_split=0.2)
```

```python
metrics(
    vixclassifier=vix_classifer, vixregressor=vix_regressor, 
    prices=True, speeches=False,
    valid=True)
```

```python
'''

hexuity.test_output writes outputs in the submission format

parameters:

    vixclassifier : classifier, default = None
        requires sklearn api

    vixregressor : regressor, default = None
        requires sklearn api

    eurclassifier : classifier, default = None
        if none : same as vix

    eurregressor : regressor, default = None
        if none : same as vix

    price_data : boolean, default = True
        decides whether to include price data 
    
    speech_data : boolean, default = True
        decides whether to include speech data

returns:

    writes a .zip file in ./answers/#attempt nbr which can be submitted to the challenge
    also writes a csv of the models score on the local test data

'''

from hexuity import test_output
```

```python
test_output(
    vixclassifier=vix_classifer, vixregressor=vix_regressor,
    price_data=True, speech_data=False)
```
